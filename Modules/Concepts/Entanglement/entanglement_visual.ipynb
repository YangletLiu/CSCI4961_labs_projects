```python

from qiskit_ibm_runtime import QiskitRuntimeService
from qiskit_ibm_runtime import SamplerV2 as Sampler
from qiskit.visualization import plot_histogram

from qiskit_aer import AerSimulator
from qiskit import QuantumCircuit
from qiskit.circuit.library import GroverOperator, MCMT, ZGate
from qiskit.visualization import plot_distribution, plot_bloch_vector

service = QiskitRuntimeService()
backend = service.backend("ibm_rensselaer")
fake_backend = AerSimulator.from_backend(backend)
```

```python
# Step 1: Create a quantum circuit with 2 qubits and 2 classical bits
qc = QuantumCircuit(2, 2)  # 2 qubits and 2 classical bits

# Step 2: Apply Hadamard gate to the first qubit
qc.h(0)

# Step 3: Apply CNOT gate (control: qubit 0, target: qubit 1)
qc.cx(0, 1)

# Step 4: Measure both qubits
qc.measure([0, 1], [0, 1])  # Correctly map qubits to classical bits
qc.draw(output="mpl")
```

```python
# Measure both qubits

simulator = AerSimulator()
job = simulator.run(qc, shots=1000) # How many copies of the circuit are executed

# Get the result and count the occurrences of each outcome
result = job.result()
counts = result.get_counts(qc)

print("Measurement results:", counts)
```
