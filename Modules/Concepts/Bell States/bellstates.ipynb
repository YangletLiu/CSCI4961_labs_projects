<h1>Bell States</h1>

<p>The Bell states are four states that can be created when two qubits are maximally entangled. The four states are:</p>

1. [$\ket{\Phi^+}$](#1)
2. [$\ket{\Phi^-}$](#2)
3. [$\ket{\Psi^+}$](#3)
4. [$\ket{\Psi^-}$](#4)

<a id="1"></a>
<h2>1. $\ket{\Phi^+}$ State</h2>
<h3> $\ket{\Phi^+} = \frac{\ket{00} + \ket{11}}{\sqrt{2}}$ <h3>
